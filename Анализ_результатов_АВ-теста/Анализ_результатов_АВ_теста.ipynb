Для изучения данных в таблицах напишем функцию, которая будет выводить фрагмент таблицы и общую информацию о ней. Дополнительно заложим в функцию приведение наименований колонок в "змеиный_регистр" и проверку на дубликаты:

```python
# функция для вывода информации о таблице

def build_info(df, name): # на вход подаем датасет и его название
    # названия колонок переводим в нижний регистр, пробелы заменяем на _
    df.columns = df.columns.str.lower().str.replace(' ', '_') 
    
    print(f'Фрагмент таблицы {name}')
    display(df.head(10))
    print('*'*75)
    
    print(f'Общая информация о таблице {name}')
    print(df.info())
    print('*'*75)
    
    # проверка на дубликаты
    if df.duplicated().sum() == 0:
        print(f'В таблице {name} дубликатов строк не обнаружено')
    else:
        print(f'В таблице {name} обнаружены дубликаты строк')
    print('*'*75)
```

## Приоритизация гипотез

### Загрузка и обзор данных

Загрузим и изучим файл с гипотезами:

```python
# загрузка датасета (полный путь не приводится, т.к. датасет не размещен в открытом доступе)
hypothesis = pd.read_csv(path1)

build_info(hypothesis, 'hypothesis')
```

Таблица включает в себя данные для приоритизации девяти гипотез, представленные в следующих столбцах:
* `hypothesis` — краткое описание гипотезы;
* `reach` — охват пользователей по 10-балльной шкале;
* `impact` — влияние на пользователей по 10-балльной шкале;
* `confidence` — уверенность в гипотезе по 10-балльной шкале;
* `efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.   

Изменения типов данных в таблице для осуществления дальнейших расчетов не требуется.

### Применение фреймворка ICE

Для удобства дальнейшего описания введем в таблицу с гипотезами новый столбец с кратким описанием гипотезы:

```python
hypothesis['hypothesis_short'] = ['+2 канала трафика', 'Запуск службы доставки',
                                  'Новые блоки рекомендаций', 'Новая структура категорий',
                                  'Новый цвет фона', 'Новая страница отзывов',
                                  'Баннеры акций на гл.странице', 'Форма подписки на осн.страницах',
                                  'Скидка в ДР']
```

Произведем расчет ICE:

```python
hypothesis['ICE'] = (hypothesis['impact'] * hypothesis['confidence']) / hypothesis['efforts']
```

Выведем итог приоритизации:

```python
hypothesis_ice = hypothesis[['hypothesis_short', 'ICE']].sort_values(by='ICE', ascending=False)                          

# порядок приоритетности по ICE
hypothesis_ice['ICE_priority'] = range(1, 10)

hypothesis_ice.style.format({'ICE': '{:.1f}'})
```

Метод приоритизации ICE показывает, что наиболее перспективными являются гипотезы (тройка лидеров):
* Скидка в ДР;
* +2 канала трафика;
* Форма подписки на осн.страницах.

### Применение фреймворка RICE

Произведем расчет RICE:

```python
hypothesis['RICE'] = (hypothesis['reach'] * 
                      hypothesis['impact'] * 
                      hypothesis['confidence']) / hypothesis['efforts']
```

Выведем итог приоритизации:

```python
hypothesis_rice = hypothesis[['hypothesis_short', 'RICE']].sort_values(by='RICE', ascending=False)

# порядок приоритетности по RICE
hypothesis_rice['RICE_priority'] = range(1, 10)

hypothesis_rice.style.format({'RICE': '{:.1f}'})
```

Метод приоритизации RICE показывает, что наиболее перспективными являются гипотезы:
* Форма подписки на осн.страницах;
* Новые блоки рекомендаций;
* +2 канала трафика / Баннеры акций на гл.странице.

### Сравнительный анализ ICE и RICE

Объединим данные по приоритизации гипотез и проведем сравнительный анализ:

```python
hypothesis_rice.merge(hypothesis_ice, on='hypothesis_short') \
               .merge(hypothesis, on='hypothesis_short')\
               [['hypothesis_short',
                 'RICE_priority',
                 'ICE_priority',
                 'reach',
                 'impact',
                 'confidence',
                 'efforts']]
```

Приоритизация гипотез при применении фрейморка RICE вместо ICE изменилась под влиянием параметра Reach (охват пользователей). Опишем изменения для лидеров в том и другом методах:
* первый приоритет по RICE имеет гипотеза по размещению формы подписки на основных страницах сайта - у нее наибольший охват пользователей, оцененный в 10 баллов. При этом без учета охвата эта гипотеза только третья по приоритетности;
* лидер по методу ICE - гипотеза по скидке в ДР - имеет почти максимальные Impact (влияние на пользователей) и Сonfidence (уверенность в гипотезе), оцененные в 9 баллов. Однако, её Reach (охват) логично оценен всего в 1 балл (минимальный среди остальных), что понижает приоритет этой гипотезы по методу RICE.

### Вывод по блоку "Приоритизация гипотез"

Определены приоритеты гипотез с помощью фреймворков ICE и RICE, которые меняются под влиянием параметра Reach (охват пользователей), который входит в формулу расчета RICE и отсутствует в ICE:
* по методу RICE первым приоритетом для проверки обладает гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок";
* по методу ICE - гипотеза "Запустить акцию, дающую скидку на товар в день рождения".

#### Распределение участников эксперимента по группам тестирования

По таблице `orders` проверим, все ли пользователи однозначно разнесены по группам эксперимента (нет ли пересечения):

```python
orders_a = orders[orders['group'] == 'A'] # заказы пользователей группы А
orders_b = orders[orders['group'] == 'B'] # заказы пользователей группы В


print(f"В группе А {orders_a['visitor_id'].nunique()} уникальных пользователя, \
совершивших {orders_a['transaction_id'].count()} заказов на {orders_a['revenue'].sum()} руб."
)

print(f"В группе B {orders_b['visitor_id'].nunique()} уникальных пользователей, \
совершивших {orders_b['transaction_id'].count()} заказов на {orders_b['revenue'].sum()} руб."
)

# список уникальных id пользователей группы А
list_users_a = orders_a['visitor_id'].unique()

# список уникальных id пользователей, которые есть одновременно в группах А и В
list_users_ab = orders_b[orders_b['visitor_id'].isin(list_users_a)]['visitor_id'].unique()

# таблица заказов пользователей, пересекающихся по группам А и В
orders_users_ab = orders[orders['visitor_id'].isin(list_users_ab)]

print('')
print(f"{len(list_users_ab)} пользователей отнесены одновременно как в группу А, так и в группу В:")
print(f"они оформили {orders_users_ab['transaction_id'].count()} заказ, что составляет \
{(orders_users_ab['transaction_id'].count() / orders['transaction_id'].count()):.0%} от общего кол-ва заказов"
)
```

Видим, что 58 пользователей отнесены как к группе А, так и к группе В: нарушен порядок разделения пользователей при проведении эксперимента. Удалить из дальнейшего рассмотрения заказы этих пользователей означает лишиться 15% данных для анализа - это слишком много. Кроме того, данные по посещениям нам доступны только в сводном виде, без разбивки по id клиентов, и отброс пересекающихся пользователей невозможно будет одновременно сделать сразу обеих таблицах - нарушится логика расчета конверсии.

Дополнительно посмотрим по отдельным пользователям их отнесение к той или иной группе теста:

```python
# сводная таблица с отражением кол-во заказов по пользователям, группам и датам
orders_users_ab.pivot_table(
    index=['visitor_id', 'group'],
    columns='date',
    values='transaction_id',
    aggfunc='count')[:6]
```

По фрагменту таблицы видим следующее:

### Подготовка таблицы с кумулятивными данными

Соберём агрегированные кумулятивные по дням данные о заказах:

```python
# массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()

# сбор агрегированных кумулятивных по дням данных о заказах
ordersAggregated = datesGroups.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transaction_id': 'nunique',
            'visitor_id': 'nunique',
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

ordersAggregated = ordersAggregated.rename(columns={'transaction_id': 'orders',
                                                   'visitor_id': 'buyers'})

ordersAggregated.head()
```

В строках таблицы `ordersAggregated` сохранены накопительные до выбранного дня включительно значения следующих показателей в разбивке по группам теста: кол-во заказов, кол-во покупателей, выручка.

Аналогично получим агрегированные кумулятивные по дням данные о посетителях интернет-магазина:

```python
# сбор агрегированных кумулятивных по дням данных о посещениях
visitorsAggregated = datesGroups.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg({'date': 'max', 'group': 'max', 'visitors': 'sum'}),
    axis=1,
).sort_values(by=['date', 'group'])

visitorsAggregated.head()
```

В строках таблицы `visitorsAggregated` сохранено накопительное до выбранного дня включительно количество посетителей интернет-магазина в разбивке по группам теста.

Объединим обе таблицы в одну:

```python
cumulativeData = ordersAggregated.merge(visitorsAggregated, on=['date', 'group'])
cumulativeData.head()
```

Таблицу `cumulativeData` с агрегированными кумулятивными по дням данными в разбивке по группам участников о посетителях, покупателях, кол-ву заказов и выручке будем использовать для построения графиков.

### Создание таблицы для анализа с "сырыми" данными

Создадим таблицу `data` с "сырыми" данными и опишем ее структуру:

```python
# датафрейм с ежедневным кол-вом посещений по группе А (посещений ЗА дату)
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

# датафрейм с кумулятивным числом посещений по группе А (посещений НА дату)
visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

# датафрейм с ежедневным кол-вом посещений по группе B (посещений ЗА дату)
visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

# датафрейм с кумулятивным числом посещений по группе B (посещений НА дату)
visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

# датафрейм с ежедневным кол-вом заказов и выручке по группе А (ЗА дату)
ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

# датафрейм с кумулятивным кол-вом заказов и выручке по группе А (НА дату)
ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = ['date', 'ordersCummulativeA', 'revenueCummulativeA']

# датафрейм с ежедневным кол-вом заказов и выручке по группе В (ЗА дату)
ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

# датафрейм с кумулятивным кол-вом заказов и выручке по группе В (НА дату)
ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = ['date', 'ordersCummulativeB', 'revenueCummulativeB']

# собираем итоговую таблицу для анализа:
data = (
    ordersADaily.merge(ordersBDaily, on='date', how='left')
    .merge(ordersACummulative, on='date', how='left')
    .merge(ordersBCummulative, on='date', how='left')
    .merge(visitorsADaily, on='date', how='left')
    .merge(visitorsBDaily, on='date', how='left')
    .merge(visitorsACummulative, on='date', how='left')
    .merge(visitorsBCummulative, on='date', how='left')
)
```

Ознакомимся с фрагментом таблицы `data`:

```python
data.head()
```

Структура итоговой таблицы с "сырыми" данными для анализа следующая:
* date — дата;
* ordersPerDateA — количество заказов в выбранную дату в группе A;
* revenuePerDateA — суммарная выручка в выбранную дату в группе A;
* ordersPerDateB — количество заказов в выбранную дату в группе B;
* revenuePerDateB — суммарная выручка в выбранную дату в группе B;
* ordersCummulativeA — суммарное число заказов до выбранной даты включительно в группе A;
* revenueCummulativeA — суммарная выручка до выбранной даты включительно в группе A;
* ordersCummulativeB — суммарное количество заказов до выбранной даты включительно в группе B;
* revenueCummulativeB — суммарная выручка до выбранной даты включительно в группе B;
* visitorsPerDateA — количество пользователей в выбранную дату в группе A;
* visitorsPerDateB — количество пользователей в выбранную дату в группе B;
* visitorsCummulativeA — количество пользователей до выбранной даты включительно в группе A;
* visitorsCummulativeB — количество пользователей до выбранной даты включительно в группе B.

Проведем проверку корректности построения таблицы, сравнив накопительные значения за последний день с суммой ежедневных значений, а также проверив соответствие суммарных значения показателей с данными исходных датасетов:

```python
print(
    data['ordersPerDateA'].sum() == data['ordersCummulativeA'].max(),
    data['ordersPerDateB'].sum() == data['ordersCummulativeB'].max(),
    data['visitorsPerDateA'].sum() == data['visitorsCummulativeA'].max(),
    data['visitorsPerDateB'].sum() == data['visitorsCummulativeB'].max(),
    data['revenuePerDateA'].sum() == data['revenueCummulativeA'].max(),
    data['revenuePerDateB'].sum() == data['revenueCummulativeB'].max(),
    data['ordersPerDateA'].sum() + data['ordersPerDateB'].sum() == orders['transaction_id'].count(),
    data['revenuePerDateA'].sum() + data['revenuePerDateB'].sum() == orders['revenue'].sum(),
    data['visitorsPerDateA'].sum() + data['visitorsPerDateB'].sum() == visitors['visitors'].sum()
)
```

Таблица `data` построена корректно.

Посчитаем статистическую значимость различия в конверсии между группами.

Создадим переменные `ordersByUsersA` и `ordersByUsersB`, в них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов:

Объявим переменные `sampleA` и `sampleB`, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули.

Переменная `sampleA` будет состоять из двух частей:
* Список с количеством заказов для каждого из пользователей: `ordersByUsersA['orders']`.
* Нули для пользователей, которые ничего не заказывали. Их количество равно разнице между суммой посетителей и количеством записей о заказах: `data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])`.

Аналогичным образом будет создана и переменная `sampleB`.

```python
# датафрейм с числом заказов для пользователей группы А
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

# датафрейм с числом заказов для пользователей группы В
sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

В `sampleA` и `sampleВ` сохранили серии данных, где каждый элемент — число заказов определённого пользователя, в том числе ноль. Число элементов в каждой серии — это количество пользователей, сумма всех элементов — количество заказов.

Найдем конверсию в каждой группе и относительный прирост конверсии группы B (конверсия группы B / конверсия группы A - 1):

```python
print(f"Конверсия в группе А по 'сырым' данным равна {sampleA.mean():.2%}")
print(f"Конверсия в группе B по 'сырым' данным равна {sampleB.mean():.2%}")

print(f"Относительный прирост конверсии группы В к группе А по 'сырым' данным равен \
{(sampleB.mean() / sampleA.mean() - 1):.1%}"
)
```

С помощью критерия Манна-Уитни, который хорошо работает на небольших выборках и устойчив к выбросам, проверим статистическую значимость различий в конверсии между группами А и В на уровне значимости 5%:
* Нулевая гипотеза: различий в конверсии между группами нет.
* Альтернативная гипотеза: различия в конверсии между группами есть.

```python
alpha = 0.05 # уровень статистической значимости

# двусторонний тест Манна-Уитни для независимых выборок
results = st.mannwhitneyu(sampleA, sampleB, alternative='two-sided')

print(f"P-value равно {results.pvalue:.5f}")

if results.pvalue < alpha:
    print(f"Отвергаем нулевую гипотезу на уровне значимости {alpha}")
else:
    print(f"Не получилось отвергнуть нулевую гипотезу на уровне значимости {alpha}")
```

**Вывод**:

P-value меньше 0.05, значит нулевую гипотезу о равенстве конверсий между группами отвергаем: анализ "сырых" данных сообщает, что в конверсии между группами есть статистически значимые различия. Относительный прирост конверсии группы В к группе А равен 13.8%.

#### Различия в среднем чеке между группами по "сырым" данным

Найдем размер среднего чека в каждой группе и относительные различия в среднем чеке между группами:

```python
print(f"Средний чек в группе А по 'сырым' данным равен \
{orders[orders['group']=='A']['revenue'].mean():.2f} руб."
)

print(f"Средний чек в группе B по 'сырым' данным равен \
{orders[orders['group']=='B']['revenue'].mean():.2f} руб."
)

print(f"Относительный прирост среднего чека группы В к группе А по 'сырым' данным равен \
{(orders[orders['group']=='B']['revenue'].mean() / orders[orders['group']=='A']['revenue'].mean() - 1):.1%}"
)
```

Проверим статистическую значимость различий в среднем чеке между сегментами, уровень значимости alpha=0.05.
* Нулевая гипотеза: различий в среднем чеке между группами нет.
* Альтернативная гипотеза: различия в среднем чеке между группами есть.

```python
alpha = 0.05 # уровень статистической значимости

# двусторонний тест Манна-Уитни для независимых выборок
results = st.mannwhitneyu(
    orders[orders['group']=='A']['revenue'],
    orders[orders['group']=='B']['revenue'],
    alternative='two-sided'
)

print(f"P-value равно {results.pvalue:.5f}")

if results.pvalue < alpha:
    print(f"Отвергаем нулевую гипотезу на уровне значимости {alpha}")
else:
    print(f"Не получилось отвергнуть нулевую гипотезу на уровне значимости {alpha}")
```

**Вывод**:

P-value больше 0.05 - статистически значимых различий в среднем чеке между группами А и В по "сырым" данным не обнаружено даже несмотря на то, что расчетный относительный прирост среднего чека группы В к группе А равен 25.9%.

### Очистка данных: устранение аномальных значений

Ранее мы рассчитали следующие показатели для отсечения аномальных значений:
* 95-й и 99-й перцентили числа заказов на одного пользователя равны 2 и 4 заказам на пользователя.
* 95-й и 99-й перцентили средних чеков равны 28000 и 58233.2 рублям

Примем за аномальных пользователей тех, кто совершил 4 и более заказов, или совершил заказ на сумму от 58 тыс.руб. (будем ориентироваться на 99-й перцентиль). Так мы уберём до 1% пользователей с наибольшим числом заказов и до 1% заказов с наибольшей стоимостью.

Сделаем срезы пользователей с числом заказов от 4 — `usersWithManyOrders` и пользователей, совершивших заказы от 58 тыс.руб. — `usersWithExpensiveOrders`. Объединим их в таблице `abnormalUsers`, удалим дубликаты, отсортируем:

```python
# срез аномальных по числу заказов пользователей 
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] >= np.percentile(ordersByUsers['orders'], 99)]['user_id'],
        ordersByUsersB[ordersByUsersB['orders'] >= np.percentile(ordersByUsers['orders'], 99)]['user_id'],
    ],
    axis=0,
)

# срез аномальных по среднему чеку пользователей 
usersWithExpensiveOrders = orders[orders['revenue'] >= np.percentile(orders['revenue'], 99)]['visitor_id']

# общий список аномальных пользователей (их id)
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)

# вывод кол-ва аномальных пользователей
print(abnormalUsers.shape[0])
```

Получили 20 аномальных пользователей.

Узнаем, как их действия повлияли на результаты теста. Посчитаем статистическую значимость различий в конверсии/среднем чеке между группами теста по очищенным данным.

### Расчет статистической значимости различий между группами по "очищенным" данным

#### Различия в конверсии между группами по "очищенным" данным

Сначала подготовим выборки количества заказов по пользователям по группам теста: создадим переменные sampleAFiltered и sampleBFiltered, в которых сохраним очищенные данные о заказах — не включая аномальных пользователей:

```python
# датафрейм с числом заказов для пользователей группы А, очищенный от аномалий
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['user_id'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

# датафрейм с числом заказов для пользователей группы А, очищенный от аномалий
sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['user_id'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

Применим статистический критерий Манна-Уитни к полученным выборкам: выведем p-value и относительное различие в конверсии между группами:

```python
print(f"Конверсия в группе А по 'очищенным' данным равна {sampleAFiltered.mean():.2%}")
print(f"Конверсия в группе B по 'очищенным' данным равна {sampleBFiltered.mean():.2%}")

print(f"Относительный прирост конверсии группы В к группе А по 'очищенным' данным равен \
{(sampleBFiltered.mean() / sampleAFiltered.mean() - 1):.1%}"
)

alpha = 0.05 # уровень статистической значимости

# двусторонний тест Манна-Уитни для независимых выборок
results = st.mannwhitneyu(sampleAFiltered, sampleBFiltered, alternative='two-sided')

print('')

print(f"P-value равно {results.pvalue:.5f}")

if results.pvalue < alpha:
    print(f"Отвергаем нулевую гипотезу на уровне значимости {alpha}")
else:
    print(f"Не получилось отвергнуть нулевую гипотезу на уровне значимости {alpha}")
```

**Вывод:** Как и в случае с "сырыми" данными, статистическая значимость достигнута: сегмент В по конверсии значительно лучше сегмента А, относительный прирост конверсии группы В к группе А составляет 13.8% (по "сырым" данным).

#### Различия в среднем чеке между группами по "очищенным" данным

Теперь повторим процедуру для среднего чека по очищенным данным: посчитаем статистическую значимость различия средних чеков после удаления аномальных пользователей.

```python
a = orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitor_id'].isin(abnormalUsers)),
            )
        ]['revenue']

b = orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitor_id'].isin(abnormalUsers)),
            )
        ]['revenue']

print(f"Средний чек в группе А по 'очищенным' данным равен {a.mean():.2f} руб.")

print(f"Средний чек в группе B по 'очищенным' данным равен {b.mean():.2f} руб.")

print(f"Относительный прирост среднего чека группы В к группе А по 'очищенным' данным равен \
{(b.mean() / a.mean() - 1):.1%}"
)

print('')

alpha = 0.05 # уровень статистической значимости

# двусторонний тест Манна-Уитни для независимых выборок
results = st.mannwhitneyu(a, b, alternative='two-sided')

print(f"P-value равно {results.pvalue:.5f}")

if results.pvalue < alpha:
    print(f"Отвергаем нулевую гипотезу на уровне значимости {alpha}")
else:
    print(f"Не получилось отвергнуть нулевую гипотезу на уровне значимости {alpha}")
```

**Вывод**: Как и в случае с "сырыми" данными, статистическая значимость не достигнута: сегмент В по среднему чеку статистически значимо не отличается от сегмента А. Однако, относительный прирост среднего чека группы В к группе А изменился по сравнению с "сырыми" данными: теперь он стал отрицательным, т.е. выбросы сильно влияли на средний чек в группе В.

### Выводы из A/B-теста
