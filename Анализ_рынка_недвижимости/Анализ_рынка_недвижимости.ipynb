**Пропуски в столбцах, сформированных на основании картографических данных**

Создадим маску для поиска пропусков в столбцах с картографическими данными (пропуски одновременно во всех столбцах):

```python
# маска для логической индексации
mask =  df['airports_nearest'].isna() & \
        df['cityCenters_nearest'].isna() & \
        df['parks_nearest'].isna() & \
        df['ponds_nearest'].isna() & \
        df['parks_around3000'].isna() & \
        df['ponds_around3000'].isna()
```

Отфильтруем таблицу по заданной маске и посмотрим распределение пропусков в привязке к населенным пунктам (ограничимся топ-10):

```python
# индексируем df по маске, смотрим топ-10 населенных пунктов
df.loc[mask, 'locality_name'].value_counts().head(10)
```

Перечисленные населенные пункты не входят в административную единицу "город Санкт-Петербург". Таким образом, предполагаю, что пропуски в данных, в основном, обусловлены отсутствием картографической привязки части населенных пунктов к объекту "город Санкт-Петербург", и их заполнение не имеет смысла.

Посмотрим еще раз на количество пропусков в столбцах с картографическими данными:

```python
# подсчет кол-ва пропусков в столбцах с картографическими данными
df[['airports_nearest', 'cityCenters_nearest',
    'parks_nearest', 'ponds_nearest',
    'parks_around3000', 'ponds_around3000']
  ].isna().sum().sort_values(ascending=False)
```

Видим, что пропусков в столбцах с расстояниями до парков/водоемов значительно больше, чем пропусков в столбцах с количеством парков/водоемов в радиусе 3000 м. Можно предположить, что в части объявлений столбец с наличием поблизости парка/водоема заполнен, при этом расстояние до объектов отсутствует. Проверим это предположение:

```python
# логическая индексация df, вывод фрагмента
df.loc[df['parks_nearest'].isna() & ~(df['parks_around3000'].isna())][['parks_nearest', 'parks_around3000']].head(5)
```

Видим, что при пропусках в столбцах с расстоянием до парка столбец количества парков заполнен нулями. Посмотрим, есть ли в этом столбце еще какие-либо значения, кроме нулей:

```python
# логическая индексация df, вывод уникальных значений столбца
df.loc[df['parks_nearest'].isna() & ~(df['parks_around3000'].isna())]['parks_around3000'].nunique()
```

Таким образом, если рядом с объектом недвижимости нет пруда, то столбец с их количеством заполнен нулями (0 - единственное уникальное значение), при этом логично, что расстояние до парка рассчитать невозможно. Отсюда появляется пропуск в столбце с расстоянием до пруда, заполнять который не имеет смысла.

Аналогично объясняется и появление пропусков в столбце с расстоянием до водоема: столбец с количеством водоемов поблизости в этом случае заполнен нулями.

*Также хочу отметить, что пропуски в картографических данных наблюдались не только по населенным пунктам, не входящим в г.Санкт-Петербург, но и в частности по объектам недвижимости, находящимся в самом Санкт-Петербурге.
```python
# контроль устранения пропусков в столбце
df['is_apartment'].isna().sum()
```

Для этих данных не могу предложить адекватный способ заполнения пропусков, поэтому оставлю всё как есть.

### Устранение дубликатов

Проверим столбец `locality_name` с категориальными переменными на наличие неявных дубликатов.

```python
# подсчет уникальных значений столбца
df['locality_name'].nunique()
```

```python
# вывод уникальных значений столбца
df['locality_name'].unique()
```

```python
# делаем построчную частичную замену данных в столбце
df['locality_name'] = df['locality_name'].str.replace('посёлок', 'поселок')
```

Примеры: "поселок при железнодорожной станции Вещево" и "поселок станции Вещево", "Мурино" и "поселок Мурино". Для дальнейшего анализа желательно избавиться и от этих дубликатов: можно будет четче выводить агрегирующие значения при группировке данных по населенным пунктам, проще указывать подписи на графиках. Поскольку нет готового словаря для правильно определения типа населенного пункта по его наименованию (а создавать его самостоятельно очень ресурсозатратно) делаем допуск: городов и поселков/деревень с одинаковыми наименованиями в данном датасете не встречается.

Создадим функцию, которая будет забирать часть имени из названия населенного пункта с первой встретившейся заглавной буквы до конца строки:

```python
def locality_name_pure(name): # функция принимает на вход имя, которое нужно очистить
    
    if pd.isna(name): # заглушка для возможных пропусков
        return name
    
    list_index=[] # список для накопления индексов вхождения заглавных букв в имя
    for i in range(len(name)): # перебираем все индексы от 0 до длины строки (количество символов в имени)
        if name[i].isupper(): # проверяем, является ли элемент с индексом i (каждая буква слова) заглавной буквой
            list_index.append(i) # если является, то добавляем индекс элемента в список
        
    real_name = name[list_index[0]:] # определяем новое имя как часть подстроки из старой строки,
                                           # начиная с индекса первого вхождения заглавной буквы и до конца строки
    return real_name # возвращаем новое имя
```

С использованием этой функции создадим новый столбец в таблице, содержащий "чистое" имя населенного пункта:

```python
# передадим имя функции в метод apply() и применим его к столбцу 'locality_name'
df['locality_name_pure'] = df['locality_name'].apply(locality_name_pure)
```

Теперь можно заполнить пропущенные значения в столбце `ceiling_height`, содержащем высоту потолков: будем идти от медианного значения по каждому населенному пункту, как и предполагалось [ранее](#intro2):

Создадим функцию для построения гистограмм и "ящиков с усами", которую будем применять для дальнейшей визуализации распределений отдельных параметров объявлений. Для гистограммы предусмотрим возможность указания максимального значения (укорачивать "правый хвост" для улучшения визуализации), боксплоты будем строить без ограничения по набору значений:

```python
def build_picture(column, x_name, x_max, bins): # передаем столбец для анализа, имя параметра,
                                               # ограничение макс.зн-я при выводе гистограммы, кол-во корзин для гистограммы
    column = column.dropna() # устранение возможных пропусков перед построением графиков
    fig = plt.figure(figsize=(8, 5)) # размер графика
    ax1 = fig.add_subplot(2,1,1) # положение графика внутри области для рисования
    ax2 = fig.add_subplot(2,1,2)
    ax1.set_title(f'Гистограмма распределения: {x_name}') # заголовок графика
    ax2.set_title(f'Диаграмма размаха: {x_name}')
    ax1.set_xlabel(x_name) # подпись оси x
    ax2.set_xlabel(x_name)
    fig.tight_layout() # устранение наложения графиков
    ax1.hist(column, bins=bins, range=(0, x_max)) # построение гистограммы
    ax2.boxplot(column, vert=False) # построение диаграммы размаха
    ax2.set_xlim(0, x_max)
    plt.show()
```

Дополнительно создадим функцию для расчета размаха "усов" боксплота - это поможет быстро получить ориентиры для описания распределения и для возможного отсечения редких значений (выбросов): границами "усов" будут 1,5 межквартильных размаха влево/вправо от первого/третьего квартилей (плюс ограничены минимальным и максимальным значениями в наборе):

```python
def whiskers(column): # передаем столбец для анализа
    left_whisker = max(column.min(), column.quantile(0.25) - 1.5 * (column.quantile(0.75) - column.quantile(0.25)))
    right_whisker = min(column.max(), column.quantile(0.75) + 1.5 * (column.quantile(0.75) - column.quantile(0.25)))    
    return (left_whisker, right_whisker)
```

### Исследование влияния различных факторов на цену квартиры

Рассчитаем коэффициенты корреляции для различных пар показателей (для пары "цена - этаж (первый, последний, другой)" корреляцию не рассчитать, поскольку тип этажа - категориальная переменная).

```python
# вывод фрагмента матрицы корреляции для набора столбцов
df_modified[['last_price', 'price_per_sq_m', 'total_area', 'rooms', 
          'year_exposition', 'month_exposition', 'weekday_exposition',
          'cityCenters_nearest']].corr().round(1).head(2)
```

Очевидно, что цена квартиры и цена за 1 кв.м. не зависят от дня/месяца/года размещения объявления: в объявлении можно менять цену в соответствии с реалиями рынка "на сейчас", и в таблице показана именно последняя цена на момент снятия объявления с публикации.

Зависимость других пар показателей дополнительно визуализируем. Площадь и число комнат будем рассматривать в паре с общей ценой квартиры, а расстояние до центра и категорию этажа в паре с ценой за 1 кв.м. (это будет логичнее, т.к. надо отсечь влияние площади квартиры на итоговую цену).

Создадим функцию вывода диаграммы рассеяния для двух величин:

```python
def build_scatter(x, y, x_name, y_name, alpha):
    plt.figure(figsize=(8, 3)) # размер графика
    plt.scatter(x=x, y=y, alpha=alpha)
    plt.title(f'Диаграмма рассеяния: {x_name} и {y_name}')
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.show()
```

**Взаимосвязь: "Цена квартиры - Общая площадь"**

```python
# вызов функции для построения диаграммы рассеяния
build_scatter(df_modified['total_area'], df_modified['last_price']/1000000, 'Общая площадь (кв.м.)', 'Цена (млн.руб.)', 0.1)
```

Коэффициент корреляции 0.8 и диаграмма рассеяния показывают достаточно сильную прямую зависимость установленной цены от общей площади квартиры. Широкая полоса облака точек отражает различную стоимость квартир одной и той же площади в разных населенных пунктах.

**Взаимосвязь: "Цена квартиры - Число комнат"**

Поскольку число комнат является дискретной величиной, то стандартная диаграмма рассеяния будет плохо читаема. Зависимость удобнее будет визуализировать через отображение медианного значения цены для квартир для каждого числа комнат. Создадим дополнительный датафрейм:

*P.S. Здесь и далее при необходимости усреднения данных буду использовать медиану распределения для сглаживания влияния больших и малых величин в наборе*

Более простым в реализации способом отображения зависимости (код короче, доп.датафрейм не нужен) могли быть диаграммы размаха, выведенные для каждой группы квартир по числу комнат (в boxplot() передать артибут группировки by='rooms'): на них также были бы видны медианы, однако было бы много лишних шумов-выбросов - сложно для восприятия.*
