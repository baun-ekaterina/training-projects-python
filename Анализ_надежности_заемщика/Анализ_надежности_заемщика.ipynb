## Заполнение пропусков

В столбцах `days_employed` и `total_income` есть пропущенные значения. В датасете они отражены как NaN (можно проводить математические операции), поскольку тип данных в соответствующих столбцах распознается как 'float' (иначе был тип 'object').

Найдем количество пропусков и их долю в общем количестве значений:

```python
# .isna() преобразует столбец в Series из True (1 - пропуск есть) и False (0 - пропусков нет)
# .sum() посчитает сумму значений, т.е. по сути кол-во пропусков True
days_employed_null_number = df['days_employed'].isna().sum()
total_income_null_number = df['total_income'].isna().sum()

# вывод на экран с использованием f-строки
print(f'Пропусков в \'days_employed\' {days_employed_null_number} шт. ({(days_employed_null_number / len(df)):.0%})')
print(f'Пропусков в \'total_income\' {total_income_null_number} шт. ({(total_income_null_number / len(df)):.0%})')
```

Количество пропущенных значений в обоих столбцах одинаково, отсюда можно предположить, что пропуски содержатся в одних и тех же строках таблицы. Проверим это с помощью логической индексации:

```python
# создаем маску для условий индексации: ищем строки с пропусками в двух столбцах одновременно
mask = df['days_employed'].isna() & df['total_income'].isna()

# индексируем датафрейм по маске и смотрим его длину
len(df[mask])
```

Длина датафрейма, содержащего пропуски одновременно в двух столбцах, равна количеству пропусков в каждом из столбцов.
Отсюда вывод: пропуски содержатся в двух столбцах одновременно в одних и тех же строках таблицы.

Информация о трудовом стаже клиента и размере его дохода является обязательной для получения кредита в банке, поэтому необходимо понять причины появления пропусков. Они могут быть такими:
* при загрузке/выгрузке данных произошел сбой (технические причины);
* при заполнении заявки на кредит сотрудником банка данные не внесены или внесены с ошибкой (человеческий фактор).

Если предположить, что в АБС банка отлажены механизмы последконтроля введенной сотрудником информации (обязательные поля не должны быть пустыми, в числовом поле не должно быть текстовых значений и пр.), то наиболее вероятная причина пропусков - технический сбой.



Чтобы не потерять 10% строк для дальнейшего анализа, необходимо заполнить пропуски данных в столбцах `days_employed` и `total_income`. Пропуски в количественных переменных заполняют характерными значениями, которые типичны для всей выборки.

Одно из возможных решений - заполнить пропуски медианным значением, т.е. числом, которое делит упорядоченную выборку пополам: ровно половина элементов больше медианы, вторая половина - меньше медианы. Применение медианы статистически не сильно исказит выборку (минимум, максимум, медиана останутся прежними) и, в отличие от среднего значения, отсечёт возможные выбросы (слишком большие или маленькие значения дохода/стажа). Применение среднего значения в данном случае менее оправданно, поскольку в датасете предложены разные категории клиентов: усреднять стаж и доходы пенсионеров и работающих нелогично.

Выведем описательную статистику данных в столбце с пропусками `days_employed`:

```python
# получение статистической сводки о данных в столбце
df['days_employed'].describe().apply('{:.0f}'.format)
```

<a id='intro1'></a>
В столбце `days_employed` с трудовым стажем данные слишком неоднородны: минимум выражен отрицательной величиной, медиана (50-й процентиль) тоже отрицательная. При этом трудовой стаж не может быть отрицательным, значит медианное значение искажено вследствие аномалий в данных. Кроме этого, максимальное значение трудового стажа составляет 401755 единиц, по описанию данных это должны быть дни, т.е. максимум будет 401755/365 = 1100 лет. Среднее значение также искажено и составляет 63046/365 = 173 года. По этому столбцу в следующем блоке проекта более детально исследуем аномалии и заполним пропуски после их устранения.

Проверим значения в столбце с пропусками `total_income`:

```python
df['total_income'].describe().apply('{:.0f}'.format)
```

<a id='intro2'></a>
В столбце `total_income` с размером дохода данные соответствуют реальности: минимум 20 тыс.руб., максимум 2,3 млн.руб, среднее 167 тыс.руб., медиана 145 тыс.руб. Однако, по-моему мнению, заполнение пропусков общим медианным значением по столбцу - не самая хорошая идея, поскольку значения доходов у пенсионеров и работающих отличаются. Поэтому заполним пропуски медианным значением по каждой категории клиентов.

<a id='intro3'></a>
Для начала, создадим таблицу медианных значений доходов в разбивке по категориям клиентов:

```python
df_grouped_1 = df.groupby('income_type')['total_income'].median()
df_grouped_1
```

Создадим функцию для одной строки, которая будет возвращать медианное значение дохода из сгруппированного датафрейма в том случае, если обнаружит пропуск в столбце с доходом. Если пропуска нет, то функция вернет исходное (заполненное) значение дохода:

```python
def median_total_income(row): # подаем на вход строку датафрейма
    if pd.isna(row['total_income']): # если обнаружен (True) пропуск в столбце 'total_income', то
        return df_grouped_1.loc[row['income_type']] # вернем значение по соответствующему 'income_type' из вспомогат. таблицы
    return row['total_income'] # если пропуск не обнаружен, вернем текущее значение в 'total_income'
```

Перезапишем значение столбца с доходом с использованием метода .apply(), в который передадим созданную функцию и применим его к кажой строке таблицы:

```python
df['total_income'] = df.apply(median_total_income, axis=1)
```

Если функция отработала верно, то медианное значение по всем категориям клиентов должно остаться прежним. Проверим это:

```python
df.groupby('income_type')['total_income'].median() == df_grouped_1
```

Кроме того, количество значений в перезаполненном столбце с заполненными пропусками должно совпасть с общим количеством строк в таблице:

```python
df['total_income'].count() == len(df)
```

Посмотрим на итоговую статистику по столбцу с доходом во всей таблице по всем категориям клиентов:

```python
df['total_income'].describe().apply('{:.0f}'.format)
```

Медианное значение изменилось незначительно: со 145 до 143 тыс.руб., однако пропуски теперь заполнены в соответствие с категорией клиента, что более логично.

Перезаписывать нулевые значения возраста будем на медианный возраст по сегментам клиентов. Алгоритм действий будет похож на [ранее](#intro3) использовавшийся:

```python
# формируем дополнительную таблицу с медианным возрастом в каждом сегоменте клиентов
df_grouped_2 = df.groupby('income_type')['dob_years'].median()
df_grouped_2
```

```python
# создаем функцию, возвращающую медианное значение возраста, если в исходнике возраст равен 0
def median_dob_year(row): # подаем на вход строку датафрейма
    if row['dob_years'] == 0: # если возраст равен 0, то
        return df_grouped_2.loc[row['income_type']] # вернем значение по соответствующему 'income_type' из вспомогат. таблицы
    return row['dob_years'] # если возраст ненулевой, то вернем текущее значение в 'dob_years'
```

```python
# перезаписываем столбец с возрастом
df['dob_years'] = df.apply(median_dob_year, axis=1)
```

```python
# проверка корректности работы функции: медиана не должна измениться по всем сегментам клиентов
df.groupby('income_type')['dob_years'].median() == df_grouped_2
```

Тогда можно думать, что алгоритм расчета нарушен: отрицательное значение появляется в результате вычитания из даты начала трудовой деятельности даты подачи заявки на кредит, а не наоборот.

Можно взять и 16 лет (минимальный возраст трудоустройства), но выбираю 18 лет исключительно из анализа собственного окружения.

```python
# создаем функцию, возвращающую значение максимально возможного стажа для "проблемных" категорий клиентов
def refresh_days_employed(row): # подаем на вход строку датафрейма
    if row['income_type'] in ['пенсионер', 'безработный']: # если категория клиента 'пенсионер' или 'безработный'
        return (row['dob_years'] - 18)*365  # рассчитаем и вернем макс.возможное значение стажа
    return row['days_employed'] # для остальных категорий возвращаем текущее значение стажа
```

## Категоризация дохода

Присвоим категорию доходу каждого клиента в зависимости от размера дохода:

* 0–30000 — 'E';
* 30001–50000 — 'D';
* 50001–200000 — 'C';
* 200001–1000000 — 'B';
* 1000001 и выше — 'A'.

Задачу выполним через наполнение списка с помощью цикла for, потенциальные ошибки в работе цикла обработаем с помощью конструкции try...except..., полученный список трансформируем в колонку и добавим ее в датафрейм. 

*Примечание:* Первоначально я хотела использовать функцию одной строки и добавление столбца в таблицу через применение к колонке с доходом метода .apply() с переданной в него функцией. Однако в ТЗ по проекту описанный способ предлагается применить при категоризации данных о цели кредита, в связи с чем решила попробовать альтернативный способ. Возможно, он приводит к более медленному расчету результата, но он работает.

Создадим в таблице новый столбец, отражающий категорию клиента по доходу.

```python
category_list = [] # создаем пустой список

for i in range(len(df)): # создаем цикл с перебором индексов от 0 до <длина датафрейма>
    try:
        # учим цикл перебирать строки таблицы,
        if 0 <= df.loc[i, 'total_income'] <=30000: # сравнивать значение в колонке 'total_income' с диапазонами
            category = 'E' # и сохранять в переменную category обозначение категории 
        elif 30001 <= df.loc[i, 'total_income'] <=50000:
            category = 'D'
        elif 50001 <= df.loc[i, 'total_income'] <=200000:
            category = 'C'
        elif 200001 <= df.loc[i, 'total_income'] <=1000000:
            category = 'B'
        elif df.loc[i, 'total_income'] >= 1000001:
            category = 'A'
        category_list.append(category) # добавляем обозначение категории в конец списка
    except:
        category_list.append('ошибка: категория не присвоена') # значение при ошибке (отрицательный доход)
```

В результате получили список категорий, сформированный в порядке возрастания индексов df. Добавим значения из этого списка в отдельную колонку.
Новую колонку можно создать методом pd.Series, но его результат меня не устроил, т.к. новая колонка добавляется в конец df. Применила метод .insert() с указанием позиции для вставки нового столбца

```python
# добавляем новый столбец после столбца 9, метод .insert() перезаписывает исходный df, присваивание можно опустить
df.insert(9, 'total_income_category', category_list)
```

```python
# контроль результата: проверка категорий условиям категоризации
df.pivot_table(index='total_income_category', values = 'total_income', aggfunc = ['min', 'median', 'max'])
```

## Категоризация целей кредита

Необходимо создать новый столбец таблицы, который определит категории для целей кредита из предложенного списка:
* 'операции с автомобилем',
* 'операции с недвижимостью',
* 'проведение свадьбы',
* 'получение образования'

Поставленную задачу предлагается решить посредством создания собственной функции и передачи ее в метод .apply() для добавление нового столбца в датафрейм.

В столбце `purpose` содержатся следующие значения:

```python
# вывод уникальных значений столбца
df['purpose'].unique()
```

```python
def find_str(text): # задаем функцию, принимающую на вход некий text
    if text.find('авто') >= 0: # ищем неотрицательный индекс вхождения подстроки 'авто' в переданный text и т.д.
        return 'операции с автомобилем'
    if text.find('недвиж') >= 0 or text.find('жиль') >= 0:
        return 'операции с недвижимостью'
    if text.find('свадьб') >= 0:
        return 'проведение свадьбы'
    if text.find('образов') >= 0:
        return 'получение образования'
    return 'ошибка: категория не присвоена' # если ни один из вариантов подстрок не подойдет, то вернется ошибка
```

```python
# создаем новый столбец датафрейма на базе существующего столбца
df['purpose_category'] = df['purpose'].apply(find_str)
```

Проверять зависимость будем через вывод сводной таблицы, строками которой будет количество детей, столбцами - факт нарушения графика возврата, а данные будут отображать количество записей в разных комбинациях "дети - факт нарушения срока возврата" (считать количество записей можно по любому столбцу таблицы, т.к. все пропуски устранены). В эту таблицу внедряем столбец, отражающий долю клиентов, нарушивших срок возврата кредита, в общем количестве клиентов каждой группы.

```python
# строим сводную таблицу
df_pivot_children = df.pivot_table(index='children', columns='debt', values='dob_years', aggfunc='count', fill_value=0) \
                      .rename(columns={0: 'payment_on_time', 1: 'delay_in_payment'}) # даем столбцам понятные наименования

# добавляем столбец с расчетом доли
df_pivot_children['delay_in_payment_%'] = (df_pivot_children['delay_in_payment'] / \
                                          (df_pivot_children['payment_on_time'] + df_pivot_children['delay_in_payment']) \
                                          * 100) \
                                          .round(1)
    
df_pivot_children
```

* в группе клиентов с 1-3 детьми доля клиентов, нарушивших срок возврата кредитов, повышается и составляет от 8,2% до 9,5%. Самое дисциплинированное поведение демонстрируют клиенты с 3 детьми, что может быть связано с более осознанным подходом к планированию бюджета семьи;
